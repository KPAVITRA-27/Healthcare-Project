```python
#converting date(object) to date(date-time)series
df['Date of Admission']=pd.to_datetime(df['Date of Admission'])
df['Discharge Date']=pd.to_datetime(df['Discharge Date'])
```

```python
df.info()
```

```python
df['admission_day'] = pd.to_datetime(df['Date of Admission']).dt.dayofweek   #5th ->sat 6th->sun
df['discharge_day']=pd.to_datetime(df['Discharge Date']).dt.dayofweek
```

```python
df['admission_day'].value_counts()
```

Observations:
Admissions are nearly the same all days.



```python
df['discharge_day'].value_counts()
```

Observations:
More discharges on wednesday and thursday as compared to other days.

```python
# Create a box plot
data=df['Age']
plt.figure(figsize=(8, 5))
sns.boxplot(data=data)
plt.title('Box Plot of Sample Data')
plt.ylabel('Values')
plt.show()
```

Observations:
The above box plot represents that there is no outliers and the data is well distributed.
